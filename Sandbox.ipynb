```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.preprocessing import label_binarize
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
lb.fit_transform(["sport","politics","art","science","science","sport","politics","poltics","poltics"])
```

```python
testDocs = ["Hello! Welcome to my sandbox.",
            "This sandbox is a place for me to create tests.",
            "Hello World! My name is Bob Jones.",
            "This is the final line of the test documents."]
            
```

```python
cv = CountVectorizer()
tfidf_vec = TfidfVectorizer()
```

```python
cvTest = cv.fit_transform(testDocs)
cvTest.toarray()
```

```python
pd.DataFrame(data = cvTest.toarray(), columns = cv.get_feature_names_out())
```

```python
tfidfTest = tfidf_vec.fit_transform(testDocs)
pd.DataFrame(data = tfidfTest.toarray(), columns = tfidf_vec.get_feature_names_out())
```
